6.Check whether the string "Unstoppable" ends with "ble" or not? Then check whether it ends with "blr" or not?

```python
#.endswith is an inbuilt function which checks whether the word ends with the letter written nside the bracket
word1="Unstoppable"
print(word1.endswith("ble"))#this is true coz the word ends with "ble"
print(word1.endswith("blr"))#this is false coz the word doesn't end with blr
```
